```python
class Student:
    
    def __init__(self,listofnames):
        self.availablenames=listofnames

    def displayAvailablenames(self):
        print("The names we have in our list are as follows:")
        print("================================")
        for name in self.availablenames:
            print(name)

    def deleteName(self,requestedName):
        if requestedName in self.availablenames:
            print("The name you requested has now been removed.")
            self.availablenames.remove(requestedName)
        else:
            print("Sorry the name you have requested to delete, is currently not in the list.")

    def addName(self,returnedName):
        self.availablenames.append(returnedName)
        print("New Name is Added Successfully.")


class Studentsname:

    def delName(self):
        print("Enter the name you'd like to delete:")
        self.name=input()
        return self.name

    def newName(self):
        print("Enter the name you'd like to add:")
        self.name=input()
        return self.name

def call():
    student=Student(["Riya Thomas","Diya K.J","Karan","Jay Ravi","Sherya Kumari","Vihan","Aromal"]) 
    studentsname=Studentsname() 
    done=False 
    while done==False:
        print(""" ======STUDENTS NAME=======
          1. Display all available names
          2. Add a name
          3. Delete a name
          4. Exit
          """)
        try:
            choice=int(input("Enter Choice:"))
            if choice==1:
                student.displayAvailablenames()
            elif choice==2:
                student.addName(studentsname.newName())
            elif choice==3:
                student.deleteName(studentsname.delName())
            elif choice==4:
                break 
                            
        except:
            print("Please enter proper choice.")

call()
```
