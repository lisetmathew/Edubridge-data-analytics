# Iphone webscraping

#  Introduction
     Web scraping is an automated method used to extract large amounts of data from websites. The data on the websites are unstructured. Web scraping helps collect these unstructured data and store it in a structured form. There are different ways to scrape websites such as online Services, APIs or writing your own code.Here I am going  to scrap the data of Apple Iphone from Flipkart to check the rating and price for each model.

```python
import requests
from bs4 import BeautifulSoup
```

#  Reading the URL

```python
url="https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&p%5B%5D=facets.brand%255B%255D%3DApple&otracker=clp_metro_expandable_6_3.metroExpandable.METRO_EXPANDABLE_Shop%2BNow_mobile-phones-store_92RED14GXPXF_wp3&fm=neo%2Fmerchandising&iid=M_cf5f8645-eaa1-4563-93ca-312a419235c1_3.92RED14GXPXF&ppt=hp&ppn=homepage&ssid=qijzqo2u1c0000001646554791013"
```

```python
r= requests.get(url)
r.status_code
```

```python
print(r.text)
```

```python
soup = BeautifulSoup(r.content, 'lxml')
```

```python
all_tables=soup.find_all("table")
print(all_tables)
```

```python
model=soup.findAll('div',{'class':"_4rR01T"})
```

```python
model
```

```python
productname=[]
for i in model:
    p=i.text
    productname.append(p)
print(productname)    
```

```python
prices=soup.findAll("div",{"class":"_30jeq3 _1_WHN1"})
```

```python
prices
```

```python
r=soup.findAll('div',{'class':"_3LWZlK"})
rating=[]
for i in r:
    ra=i.text
    rating.append(ra)
print(rating)    
```

```python
len(rating)
```

```python
len(prices)
```

```python
len(model)
```

```python
import pandas as pd
d={'productname':productname,'price':prices,'rating':rating}
data=pd.DataFrame(d)
data
```
